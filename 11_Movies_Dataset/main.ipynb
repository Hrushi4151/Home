```python
import pandas as pd

# --------------------------------------------------
# Load Dataset
# --------------------------------------------------

df = pd.read_csv("movies_metadata.csv", low_memory=False)

# Display column names
print("Columns in Dataset:\n")
print(df.columns)

# Display first 5 rows
print("\nOriginal Dataset:\n")
print(df.head())
```

```python

# --------------------------------------------------
# m. Create data subsets for different languages
# --------------------------------------------------

english_movies = df[df['original_language'] == 'en']
french_movies = df[df['original_language'] == 'fr']
hindi_movies = df[df['original_language'] == 'hi']

print("\nEnglish Movies Subset:\n")
print(english_movies[['title', 'original_language']].head())

print("\nFrench Movies Subset:\n")
print(french_movies[['title', 'original_language']].head())

print("\nHindi Movies Subset:\n")
print(hindi_movies[['title', 'original_language']].head())
```

```python
# --------------------------------------------------
# n. Merge two subsets
# --------------------------------------------------

merged_data = pd.concat([english_movies, french_movies])

print("\nMerged Dataset (English + French):\n")
print(merged_data[['title', 'original_language']].head())
```

```python
# --------------------------------------------------
# o. Sort Data using customer ratings
# --------------------------------------------------

# Convert vote_average to numeric
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')

sorted_data = df.sort_values(by='vote_average', ascending=False)

print("\nMovies Sorted by Customer Ratings:\n")
print(sorted_data[['title', 'vote_average']].head())
```

```python
# --------------------------------------------------
# p. Transposing Data
# --------------------------------------------------

transposed_data = df.transpose()

print("\nTransposed Data")
print(transposed_data.head())
```

```python

# --------------------------------------------------
# q. Melting Data to Long Format
# --------------------------------------------------

melted_data = pd.melt(
    df,
    id_vars=['title'],
    value_vars=['popularity', 'vote_average', 'vote_count'],
    var_name='Metric',
    value_name='Value'
)

print("\nMelted Data (Long Format):\n")
print(melted_data.head())
```

```python

# --------------------------------------------------
# r. Casting Data to Wide Format
# --------------------------------------------------

wide_data = melted_data.pivot_table(
    index='title',
    columns='Metric',
    values='Value',
    aggfunc='first'
)

print("\nWide Format Data:\n")
print(wide_data.head())
```
