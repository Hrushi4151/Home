```python
import pandas as pd
```

```python
column_names=[
    'Class',
    'AGE',
    'SEX',
    'STEROID',
    'ANTIVIRALS',
    'FATIGUE',
    'MALAISE',
    'ANOREXIA',
    'LIVER BIG',
    'LIVER FIRM',
    'SPLEEN PALPABLE',
    'SPIDERS',
    'ASCITES',
    'VARICES',
    'BILIRUBIN',
    'ALK PHOSPHATE',
    'SGOT',
    'ALBUMIN',
    'PROTIME',
    'HISTOLOGY'
]
df=pd.read_csv('hepatitis.csv', header=None, names=column_names, index_col=False)
df.head()
```

```python
melted_dataset=pd.melt(
    df,
    id_vars='SEX',
    value_vars=[
        'AGE',
        'Class',
        'ANTIVIRALS'
    ],
    var_name='Attributes',
    value_name='Values'
)
melted_dataset.head()
```

```python
wide_format=melted_dataset.pivot_table(
    index='SEX',
    columns='Attributes',
    values='Values',
    aggfunc='mean'
)
wide_format.head()
```
