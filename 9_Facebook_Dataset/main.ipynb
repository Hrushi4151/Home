```python
# Import pandas
import pandas as pd

# Load Facebook dataset
df = pd.read_csv("dataset_Facebook.csv", sep=';')

# Display first 5 rows
print(df.head())


# --------------------------------------------------
# a. Create data subsets for type of post
# --------------------------------------------------

# Subset for Photo posts
photo_posts = df[df['Type'] == 'Photo']

# Subset for Video posts
video_posts = df[df['Type'] == 'Video']

print("\nPhoto Posts")
print(photo_posts.head())

print("\nVideo Posts")
print(video_posts.head())


# --------------------------------------------------
# b. Merge two subsets
# --------------------------------------------------

merged_data = pd.concat([photo_posts, video_posts])

print("\nMerged Data")
print(merged_data.head())


# --------------------------------------------------
# c. Sort Data on Page total likes
# --------------------------------------------------

sorted_data = df.sort_values(by='Page total likes')

print("\nSorted Data")
print(sorted_data.head())


# --------------------------------------------------
# d. Transposing Data
# --------------------------------------------------

transposed_data = df.transpose()

print("\nTransposed Data")
print(transposed_data.head())


# --------------------------------------------------
# e. Melting Data to long format
# --------------------------------------------------

melted_data = pd.melt(
    df,
    id_vars=['Type'],
    value_vars=['like', 'share', 'comment'],
        var_name='Metric',
    value_name='Count'
)

print("\nMelted Data")
print(melted_data.head())


# --------------------------------------------------
# f. Casting data to wide format
# --------------------------------------------------

wide_data = melted_data.pivot_table(
    index='Type',
    columns='variable',
    values='value',
    aggfunc='mean'
)

print("\nWide Format Data")
print(wide_data)
```
