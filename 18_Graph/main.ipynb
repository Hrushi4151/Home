```python
# =====================================================
# Heart Disease Dataset Visualization using Python
# =====================================================

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------------------------------
# Load Dataset
# -----------------------------------------------------

file_path = "Cleavland.csv"

# Column names
columns = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs',
    'restecg', 'thalach', 'exang', 'oldpeak',
    'slope', 'ca', 'thal', 'target'
]

# Read dataset
df = pd.read_csv(file_path, header=None, names=columns, skiprows=1)

# Convert target into 2 classes
# 0 = No Disease
# 1 = Disease

df['target'] = df['target'].apply(lambda x: 0 if x == 0 else 1)

# Display first 5 rows
print(df.head())

# =====================================================
# a. HISTOGRAM
# Objective:
# To understand distribution of Age
# =====================================================

plt.figure(figsize=(6,4))

plt.hist(df['age'], bins=10)

plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()

# =====================================================
# b. PIE CHART
# Objective:
# To show percentage of patients with and without disease
# =====================================================

heart_disease_count = df['target'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    heart_disease_count,
    labels=['No Disease', 'Disease'],
    autopct='%1.1f%%'
)

plt.title('Heart Disease Distribution')

plt.show()

# =====================================================
# c. BOX PLOT
# Objective:
# To detect outliers in Cholesterol values
# =====================================================

plt.figure(figsize=(6,4))

plt.boxplot(df['chol'])

plt.title('Box Plot of Cholesterol')
plt.ylabel('Cholesterol')

plt.show()

# =====================================================
# d. SCATTER PLOT
# Objective:
# To study relationship between Age and Cholesterol
# =====================================================

plt.figure(figsize=(6,4))

plt.scatter(df['age'], df['chol'])

plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')

plt.show()

# =====================================================
# e. SCATTER PLOT WITH BOXPLOTS
# Objective:
# To visualize relationship and distribution together
# =====================================================

fig = plt.figure(figsize=(8,8))

# Scatter Plot
ax_scatter = fig.add_axes([0.1, 0.1, 0.65, 0.65])

ax_scatter.scatter(df['age'], df['chol'])

ax_scatter.set_xlabel('Age')
ax_scatter.set_ylabel('Cholesterol')
ax_scatter.set_title('Scatter Plot with Boxplots')

# Top Boxplot
ax_box_x = fig.add_axes([0.1, 0.77, 0.65, 0.15])

ax_box_x.boxplot(df['age'], vert=False)

ax_box_x.set_xticks([])

# Right Boxplot
ax_box_y = fig.add_axes([0.77, 0.1, 0.15, 0.65])

ax_box_y.boxplot(df['chol'])

ax_box_y.set_yticks([])

plt.show()
```
