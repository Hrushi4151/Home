```python
import pandas as pd

# Define column names
columns = [
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    'income'
]

# Load Dataset
df = pd.read_csv("adult.csv", names=columns)

# Remove extra spaces from string values
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Display first rows
print("Original Dataset:\n")
print(df.head())
```

```python
# --------------------------------------------------
# s. Create data subsets for Country, Sex, Race
# --------------------------------------------------

india_data = df[df['native-country'] == 'India']
male_data = df[df['sex'] == 'Male']
white_data = df[df['race'] == 'White']

print("\nIndia Subset:\n")
print(india_data.head())

print("\nMale Subset:\n")
print(male_data.head())

print("\nWhite Race Subset:\n")
print(white_data.head())


# --------------------------------------------------
# t. Merge two subsets
# --------------------------------------------------

merged_data = pd.concat([india_data, male_data])

print("\nMerged Dataset:\n")
print(merged_data.head())

```

```python

# --------------------------------------------------
# u. Sort Data using Age
# --------------------------------------------------

sorted_data = df.sort_values(by='age', ascending=False)

print("\nSorted Data by Age:\n")
print(sorted_data[['age', 'sex', 'native-country']].head())
```

```python

# --------------------------------------------------
# v. Transposing Data
# --------------------------------------------------

transposed_data = df.transpose()

print("\nTransposed Data:\n")
print(transposed_data)
```

```python
# --------------------------------------------------
# w. Melting Data to Long Format
# --------------------------------------------------

melted_data = pd.melt(
    df,
    id_vars=['sex'],
    value_vars=['age', 'hours-per-week', 'education-num'],
    var_name='Attribute',
    value_name='Value'
)

print("\nMelted Data:\n")
print(melted_data.head())
```

```python
# --------------------------------------------------
# x. Casting Data to Wide Format
# --------------------------------------------------

wide_data = melted_data.pivot_table(
    index='sex',
    columns='Attribute',
    values='Value',
    aggfunc='mean'
)

print("\nWide Format Data:\n")
print(wide_data)
```
