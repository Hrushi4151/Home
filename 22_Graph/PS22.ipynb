```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col_names = [
    'age', 'workclass', 'fnlwgt', 'education',
    'education_num', 'marital_status', 'occupation',
    'relationship', 'race', 'sex', 'capital_gain',
    'capital_loss', 'hours_per_week', 'native_country',
    'income'
]
```

```python
df = pd.read_csv("adult.csv", names=col_names)
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Display First 5 Rows
print(df.head())
```

```python
# a. HISTOGRAM
# Objective:
# To show the distribution of age in the dataset.

plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=10)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# b. PIE CHART
# Objective:
# To show percentage distribution of males and females.

gender_count = df['sex'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count.values,
        labels=gender_count.index,
        autopct='%1.1f%%')
plt.title("Pie Chart of Gender")
plt.show()
```

```python
# c. BOX PLOT
# Objective:
# To identify spread and outliers in age data.

plt.figure(figsize=(6,4))
plt.boxplot(df['age'])
plt.title("Box Plot of Age")
plt.ylabel("Age")
plt.show()
```

```python
# d. SCATTER PLOT
# Objective:
# To show relationship between age and hours worked per week.

plt.figure(figsize=(6,4))
plt.scatter(df['age'], df['hours_per_week'])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Hours Per Week")
plt.show()
```

```python
# e. ADD BOXPLOTS TO A SCATTERPLOT
# Objective:
# To compare data distribution along with relationship
# between age and working hours.

fig, ax = plt.subplots(figsize=(7,5))

# Scatter Plot
ax.scatter(df['age'], df['hours_per_week'])

# Labels
ax.set_title("Scatter Plot with Boxplot Information")
ax.set_xlabel("Age")
ax.set_ylabel("Hours Per Week")

plt.show()

# Separate Boxplots for comparison
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.boxplot(df['age'])
plt.title("Boxplot of Age")

plt.subplot(1,2,2)
plt.boxplot(df['hours_per_week'])
plt.title("Boxplot of Hours Per Week")

plt.show()
```
