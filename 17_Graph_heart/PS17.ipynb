```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
columns = [
    'age', 'sex', 'cp', 'trestbps', 'chol',
    'fbs', 'restecg', 'thalach', 'exang',
    'oldpeak', 'slope', 'ca', 'thal', 'target'
]
```

```python
df = pd.read_csv("Cleavland.csv", names=columns)

# Display dataset
print(df.head())
```

```python
# =====================================================
# a) Histogram
# Objective: Distribution of Age
# =====================================================

plt.hist(df['age'], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# =====================================================
# b) Dot Plot
# Objective: Show cholesterol values
# =====================================================

plt.plot(df['chol'], 'o')

plt.title("Dot Plot of Cholesterol")
plt.ylabel("Cholesterol")

plt.show()
```

```python
# =====================================================
# c) Bar Plot
# Objective: Compare chest pain types
# =====================================================

cp_count = df['cp'].value_counts()

plt.bar(cp_count.index, cp_count.values)

plt.title("Bar Plot of Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")

plt.show()
```

```python
# =====================================================
# d) Line Chart
# Objective: Show resting blood pressure trend
# =====================================================

plt.plot(df['trestbps'])

plt.title("Line Chart of Resting Blood Pressure")
plt.xlabel("Patient Index")
plt.ylabel("Blood Pressure")

plt.show()
```

```python
# e) Histogram and scatter plot to box plot 
# =====================================================
# 1) Histogram added to Box Plot
# Objective:
# To visualize distribution and outliers together
# =====================================================

fig, ax = plt.subplots()

# Histogram
ax.hist(df['chol'], bins=10)

# Boxplot added on top
ax.boxplot(df['chol'], vert=False)

plt.title("Histogram with Box Plot")

plt.show()
```

```python
# =====================================================
# 2) Scatter Plot added to Box Plot
# Objective:
# To visualize data spread and relationship together
# =====================================================

fig, ax = plt.subplots()

# Scatter Plot
ax.scatter(df['age'], df['thalach'])

# Box Plot added
ax.boxplot(df['thalach'], positions=[200])

plt.title("Scatter Plot with Box Plot")

plt.xlabel("Age")
plt.ylabel("Thalach")

plt.show()
```

```python
# =====================================================
# f) Pie Chart
# Objective: Heart disease proportion
# =====================================================

target_count = df['target'].value_counts()

plt.pie(target_count.values,
        labels=target_count.index,
        autopct='%1.1f%%')

plt.title("Pie Chart of Heart Disease")

plt.show()
```

```python
# =====================================================
# g) Box Plot
# Objective: Detect outliers
# =====================================================

plt.boxplot(df['chol'])

plt.title("Box Plot of Cholesterol")

plt.show()
```

```python
# =====================================================
# h) Scatter Plot
# Objective: Relation between Age and Heart Rate
# =====================================================

plt.scatter(df['age'], df['thalach'])

plt.title("Scatter Plot of Age vs Max Heart Rate")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")

plt.show()
```

```python
# =====================================================
# i) Box Plot added to Scatter Plot
# Objective: Spread with individual values
# =====================================================

plt.boxplot(df['thalach'])

plt.scatter([1]*len(df['thalach']), df['thalach'])

plt.title("Box Plot with Scatter Plot")
plt.ylabel("Max Heart Rate")

plt.show()
```
