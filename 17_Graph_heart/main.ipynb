```python
# ==========================================
# Heart Disease Dataset Visualization
# ==========================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
# Replace 'heart.csv' with your dataset file name
columns = [
    'age', 'sex', 'cp', 'trestbps', 'chol',
    'fbs', 'restecg', 'thalach', 'exang',
    'oldpeak', 'slope', 'ca', 'thal', 'target'
]

df = pd.read_csv("Cleavland.csv", names=columns)

# Display first 5 rows
print(df.head())
# ==========================================
# a. HISTOGRAM
# Objective:
# To show the distribution of Age in the dataset
# ==========================================

plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=10)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()


# ==========================================
# b. DOT PLOT
# Objective:
# To show individual cholesterol values
# ==========================================

plt.figure(figsize=(8,2))
plt.plot(df['chol'], ['o'] * len(df), 'o')
plt.title("Dot Plot of Cholesterol")
plt.xlabel("Cholesterol")
plt.yticks([])
plt.show()


# ==========================================
# c. BAR PLOT
# Objective:
# To compare number of male and female patients
# sex: 1 = Male, 0 = Female
# ==========================================

sex_count = df['sex'].value_counts()

plt.figure(figsize=(5,4))
plt.bar(['Male', 'Female'], sex_count)
plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()


# ==========================================
# d. LINE CHART
# Objective:
# To show variation of resting blood pressure
# ==========================================

plt.figure(figsize=(8,4))
plt.plot(df['trestbps'])
plt.title("Resting Blood Pressure Line Chart")
plt.xlabel("Index")
plt.ylabel("Resting BP")
plt.show()


# ==========================================
# e. BOX PLOT WITH HISTOGRAM AND SCATTER
# Objective:
# To identify spread and outliers in cholesterol
# ==========================================

# BOX PLOT
plt.figure(figsize=(6,4))
plt.boxplot(df['chol'])
plt.title("Box Plot of Cholesterol")
plt.ylabel("Cholesterol")
plt.show()

# HISTOGRAM
plt.figure(figsize=(6,4))
plt.hist(df['chol'], bins=10)
plt.title("Histogram of Cholesterol")
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.show()

# SCATTER PLOT
plt.figure(figsize=(6,4))
plt.scatter(df['age'], df['chol'])
plt.title("Scatter Plot: Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```
