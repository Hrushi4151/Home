```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col_names = [
    'age', 'workclass', 'fnlwgt', 'education',
    'education_num', 'marital_status', 'occupation',
    'relationship', 'race', 'sex', 'capital_gain',
    'capital_loss', 'hours_per_week', 'native_country',
    'income'
]
```

```python
df = pd.read_csv('adult.csv',names=col_names)
df.head()
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
print(df.head())
```

```python

# f. HISTOGRAM
# To show the distribution of age in the dataset.

plt.figure(figsize=(6,4))
plt.hist(df['age'],bins=10)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# DOT PLOT
# To visualize individual age values.

plt.figure(figsize=(8,2))
plt.plot(df['age'],[1]*len(df['age']),'o')
plt.title("Dot Plot of Age")
plt.xlabel("Age")
plt.yticks([])
plt.show()
```

```python
# BAR PLOT
# To show count of males and females.

gender_count = df['sex'].value_counts()
plt.figure(figsize=(5,4))

plt.bar(gender_count.index,gender_count.values)
plt.title("Bar plot of gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Line Chart
# To show age values in sequence.

plt.figure(figsize=(8,4))
plt.plot(df['age'].head(50))
plt.title("Line Chart of Age")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

```python
# BOX PLOT + HISTOGRAM + SCATTER PLOT
# To identify spread and outliers in age data.

# BOX PLOT
plt.figure(figsize=(6,4))
plt.boxplot(df['age'])
plt.title("Box Plot of Age")
plt.ylabel("Age")
plt.show()


# SCATTER PLOT
plt.figure(figsize=(6,4))
plt.scatter(df.index, df['age'])
plt.title("Scatter Plot of Age")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()


# HISTOGRAM
plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=10)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
