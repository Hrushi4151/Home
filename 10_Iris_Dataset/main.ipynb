```python
import pandas as pd

# Define column names
columns = [
    'SepalLengthCm',
    'SepalWidthCm',
    'PetalLengthCm',
    'PetalWidthCm',
    'Species'
]

# Load dataset
df = pd.read_csv("Iris.csv", names=columns)

print("Original Dataset:\n")
print(df.head())
```

```python

# --------------------------------------------------
# g. Create data subsets for different species
# --------------------------------------------------

setosa = df[df['Species'] == 'Iris-setosa']
versicolor = df[df['Species'] == 'Iris-versicolor']
virginica = df[df['Species'] == 'Iris-virginica']

print("\nSetosa Subset:\n")
print(setosa.head())

print("\nVersicolor Subset:\n")
print(versicolor.head())

print("\nVirginica Subset:\n")
print(virginica.head())
```

```python
# --------------------------------------------------
# h. Merge two subsets
# --------------------------------------------------

merged_data = pd.concat([setosa, versicolor])

print("\nMerged Dataset:\n")
print(merged_data.head())
```

```python

# --------------------------------------------------
# i. Sort Data on Petal Length
# --------------------------------------------------

sorted_data = df.sort_values(by='PetalLengthCm', ascending=False)

print("\nSorted Data:\n")
print(sorted_data.head())
```

```python
# --------------------------------------------------
# j. Transposing Data
# --------------------------------------------------

transposed_data = df.transpose()

print("\nTransposed Data")
print(transposed_data.head())
```

```python
# --------------------------------------------------
# k. Melting Data to Long Format
# --------------------------------------------------

melted_data = pd.melt(
    df,
    id_vars=['Species'],
    value_vars=[
        'SepalLengthCm',
        'SepalWidthCm',
        'PetalLengthCm',
        'PetalWidthCm'
    ],
    var_name='Measurement',
    value_name='Value'
)

print("\nMelted Data:\n")
print(melted_data.head())
```

```python
# --------------------------------------------------
# l. Casting Data to Wide Format
# --------------------------------------------------

wide_data = melted_data.pivot_table(
    index='Species',
    columns='Measurement',
    values='Value',
    aggfunc='mean'
)

print("\nWide Format Data:\n")
print(wide_data)
```
